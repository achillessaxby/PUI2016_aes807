```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline

df = pd.DataFrame(data[["Block","Job #"]], columns=['Block','Job #'])

df.plot.scatter(x='Block', y='Job #', color='DarkBlue', label='Block v/s Job #');

plt.title('NYCOpenData Plotting', fontsize = 15 , color='red')

plt.text(0,-150000000,'Figure : Assignment 3 Plotting', fontsize = 10, fontweight = 'bold')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline

df = pd.DataFrame(data[["Block","Lot"]], columns=['Block','Lot'])

df.plot.scatter(x='Block', y='Lot', color='DarkBlue', label='Block v/s Lot');

plt.title('NYCOpenData Plotting', fontsize = 15 , color='red')

plt.text(-20000,-5000,'Figure : Assignment 3 Plotting', fontsize = 10, fontweight = 'bold')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline

df = pd.DataFrame(data[["Job #","Bin #"]], columns=['Job #','Bin #'])

df.plot.scatter(x='Job #', y='Bin #', color='DarkBlue', label='Job v/s Bin #');

plt.title('NYCOpenData Plotting', fontsize = 15 , color='red')

plt.text(0,-1500000,'Figure : Assignment 3 Plotting', fontsize = 10, fontweight = 'bold')

plt.legend(loc='upper left')
plt.show()
```
